```python
##  write the way

geography=pd.get_dummies(x['Geography'],dtype=int,drop_first=True)
gender=pd.get_dummies(x["Gender"],dtype=int,drop_first=True)
```

```python
## concatenate these variables with dataframe

x.drop(["Geography","Gender"],inplace=True,axis=1)
```

```python
x=pd.concat([x,geography,gender],axis=1)
```

```python
x.head()
```

```python
##  spliting the dataset into training set and test set

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
## part 2 is create the ANN

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
## lets initialize the ANN
classifier=Sequential()
```

```python
## adding the input layer

classifier.add(Dense(units=11,activation='relu'))
```

```python
## adding first hidden layer

classifier.add(Dense(units=7,activation='relu'))
classifier.add(Dropout(0.2))
```

```python
## adding second hidden layer

classifier.add(Dense(units=6,activation='relu'))
classifier.add(Dropout(0.3))
```

```python
## adding the output layer

classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
## suppose if we want to fix manual learning rate

import tensorflow as tf
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
## early stopping

import tensorflow as tf

early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
## summarize history for accuracy

plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','validation'],loc='upper left')
plt.show()
```

```python
## summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','validation'],loc='upper left')
plt.show()
```

```python
## part 3 - making predictions and evaluating the model
# predicting the test set results
y_pred=classifier.predict(x_test)
y_pred=(y_pred>=0.5)
```

```python
## make the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
## calculate the accuracy

from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```
